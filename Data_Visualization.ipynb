```python
import pandas as pd

import matplotlib.pyplot as plt

data = {'date': ['2023-08-01','2023-08-02','2023-08-03'],
        'price': [10,20,30]}

df= pd.DataFrame(data)

df['date'] = pd.to_datetime(df['date'])

df.plot(x = 'date', y= 'price', kind='line')
plt.xlabel('date')
plt.ylabel('price')

plt.show()
```

```python
data = {'Product': ['A','B','C'],
        'Sales': [10,20,30]}

df = pd.DataFrame(data)

df.plot(x= 'Product', y= 'Sales', kind='bar')
plt.xlabel('Product')
plt.ylabel('Sales')

plt.show()
```

```python
data = {'Hour_studied': [2,4,5,7,8,10],
        'Exam_scores': [60,40,50,70,80, 60]}

df = pd.DataFrame(data)

df.plot(x= 'Hour_studied', y= 'Exam_scores', kind= 'scatter')
plt.show()
```

```python
data = {'Hour_studied': [2,4,5,7,8,10],
        'Exam_scores': [60,40,50,70,80, 60]}

df = pd.DataFrame(data)

df.plot(x= 'Hour_studied', y= 'Exam_scores', kind= 'pie')
plt.show()
```
