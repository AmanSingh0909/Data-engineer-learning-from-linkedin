```python
import numpy as np

arr1 = np.array([5,4,8,9,7])
arr2 = np.array([1,2,3,4,10])

sum = arr1 + arr2
print(sum)
```

```python
sum = arr1 - arr2
print(sum)
```

```python
sum = arr1 * arr2
print(sum)
```

```python
sum = arr1 / arr2
print(sum)
```

```python
mean = np.mean(arr1)
print(mean)
```
