```python
class Employee:
  def __init__(self, name, position, salary):
    self.name = name
    self.position = position
    self.salary = salary

  def display_info(self):
    return f"{self.name} is a {self.position} earning ${self.salary} Annually"
```

```python
employee1 = Employee("John","Manager",50000)
employee2 = Employee("Aman", "Developer", 60000)
```

```python
print(employee1.display_info())
```

```python
print(employee2.display_info() )
```

```python
employee2.salary = 100000
print(employee2.display_info())
```
