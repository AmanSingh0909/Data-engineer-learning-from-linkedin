```python
import numpy as np

arr1 = np.array([[1,2],[3,4]])
arr2 = np.array([[5,6],[7,8]])

mul = np.dot(arr1, arr2)
print(mul)
```

```python
invr = np.linalg.inv(arr1)
print(invr)
```
