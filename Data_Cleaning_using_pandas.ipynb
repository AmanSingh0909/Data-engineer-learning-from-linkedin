```python
import pandas as pd

df = pd.read_csv('/content/Student_Performance_Dataset.csv')
```

```python
drop_df = df.dropna()
print(drop_df.to_string())
```

```python
df.fillna(100, inplace= True)
```

```python
df['Age'].fillna(25)
```

```python
df_new = df.drop_duplicates()
```

```python
df['Previous_Year_Score'] = pd.to_datetime(df['Previous_Year_Score'], format = '%d%m%y', errors= 'ignore')
```

```python
avg_gender_math_score = drop_df.groupby('Gender')['Science_Score'].mean()
print(avg_gender_math_score)
```

```python
drop_df.aggregate(['sum','min'])
```

```python
new_order_file = pd.read_csv('')
merged_data = pd.merge(drop_df, new_order_file, on='Student_ID')
print(merged_data)
```
